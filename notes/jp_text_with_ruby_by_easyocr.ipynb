# ルビ付きの日本語文章に対するORC by EasyOCR
- https://github.com/JaidedAI/EasyOCR
- [【Python】画像中の文字認識の簡単実装【EasyOCR】](https://plantprogramer.com/easyocr/)

```python
import easyocr
import cv2

reader = easyocr.Reader(['ja','en']) # this needs to run only once to load the model into memory
```

```python
file_path = "../data/kokoro-ruby-2.png"

result = reader.readtext(file_path)

print(result)
```

```python
simple_result = reader.readtext(file_path, detail=0)

print(simple_result)
```

```python
image = cv2.imread(file_path)

for (bbox, text, prob) in result:
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = (int(top_left[0]), int(top_left[1]))
    bottom_right = (int(bottom_right[0]), int(bottom_right[1]))

    cv2.rectangle(image, top_left, bottom_right, (0, 0, 255), 2)

cv2.imwrite('../data/result_easyocr_bounding_box.jpg', image)
```
