# ルビなしの日本語文章に対するORC by EasyOCR
- https://github.com/JaidedAI/EasyOCR
- [【Python】画像中の文字認識の簡単実装【EasyOCR】](https://plantprogramer.com/easyocr/)

```python
import easyocr
import cv2

reader = easyocr.Reader(['en', 'ja']) # this needs to run only once to load the model into memory
```

```python
file_path = "../data/kokoro-without-ruby-2.png"

result = reader.readtext(file_path)

print(result)
```

```python
simple_result = reader.readtext(file_path, detail=0)

print(simple_result)
```
